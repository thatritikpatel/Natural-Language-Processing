```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("khulasasndh/game-of-thrones-books")

print("Path to dataset files:", path)
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)


import os
for dirname, _, filenames in os.walk(path):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
import gensim
import os
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for filename in os.listdir(path):
    f = open(os.path.join(path,filename),encoding='Latin')
    corpus = f.read()
    token = sent_tokenize(corpus)
    for sent in token:
        story.append(simple_preprocess(sent))
```

```python
import nltk
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop = stopwords.words('english')
```

```python
def remove_stopword(text):
    return [[item for item in sentence if isinstance(item,str) and  item.lower() not in stop]for sentence in text]
story = remove_stopword(story)
```

```python
from gensim.models import Word2Vec
word2vec = Word2Vec(vector_size = 100,window = 5)
```

```python
word2vec.build_vocab(story)
word2vec.train(story,total_examples=word2vec.corpus_count,epochs = word2vec.epochs)
```
