# **What are regular expressions?**

The Regex or Regular Expression is a way to define a pattern for searching or manipulating strings. We can use a regular expression to match, search, replace, and manipulate inside textual data.

Regular expressions are instrumental in extracting information from text such as log files, spreadsheets, or even textual documents.

For example, Below are some of the cases where regular expressions can help you to save a lot of time.

1. Searching and replacing text in files
2. Validating text input, such as password and email address
3. Rename a hundred files at a time. For example, You can change the extension of all files using a regex pattern

we used the re.findall() method to match our pattern.

```python
# import RE module
import re

data = "My Enrollment ID is 32"
res = re.findall(r"\d", data)
# extract mathing value
print(res)
```

```python
"""
Methods in re package:
split - splits the string according to pattern,
findall  - returns all matches,
match  - checks only the beginning of the string,
search  - returns the first occurrence
sub  - returns a new string with occurrences of the pattern replaced with the supplied string
"""
```

```python
text = 'Take one up 1-3-2019 on 23 idea.One idea 45 at a Time 12-11-2020'
```

```python
re.findall(pattern="([0-9]{1,2}-[0-9]{1,2}-[0-9]{4})", string=text)
```

```python
re.findall(pattern="([0-9]{1,2}-[0-9]{1,2}-[0-9]{4})$", string=text)
```

```python
str = 'Take one up 1-3-2019 on 23 idea.One idea 45 at a Time 12-11-2020'
re.findall(pattern="[0-9]{1,2}-[0-9]{1,2}-[0-9]{4}$", string=str) #the ending of the string should be this character [0-9]{4}
```

```python
import re

# Example string
text = "Hello, world!"

# Perform a search for the word "wor"
match = re.search(r"wor", text)

# Print the match information
print(match.span())  # Output: (7, 10)
print(text[7:10])    # Output: "wor"
```

Demonstrating the importance of escabe symbol. If we don't want to put the escape symbol in order to escape the special meaning, make it a raw string.

```python
import re
n = re.match("(\d+)\.(\d+)", "24.1632")
print(n.group())

m = re.match(r"(\d+).(\d+)", "24.1632")
print(m.group())
```

```python
#splits the string into multiple strings where-ever there is the specified delimiter
print(str)
result=re.split('\d+',str)  #here we are using 1 or more digit as the delimiter
print(result)
```

```python
#sub function will substitute x pattern with y pattern in the given string
result = re.sub('One', 'Two', str)
print(result)
```

```python
#sub function will substitute x pattern with y pattern in the given string
result = re.sub('(O|o)ne', 'Two', str)
print(result)
```
